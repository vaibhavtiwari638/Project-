# This notebook demonstrates the working of basic functionalities in Pandas library for data analysis tasks.

# Data Structures in Pandas :
1. Series = one dimensional labled data.
2. Data Frames = 2D datasets.
3. Pannel = 3D form of data.

```python
import pandas as pd
```

#Series data Structure

```python
list =[1,2,3,4,5,6]
s = pd.Series(list)
print(s)
```

```python
s = pd.Series(list,index=('a','s','d','r','p','w'),dtype= float)
print(s)
```

```python
dict ={
    "shapes":["square","triangle","rectangle"],
    "size":["large","rigid","solid"],
    "rank":[1,2,4]
}
var = pd.Series(dict)
print(dict)
```

# Data Frames

```python
l =[1,2,3,4,5,6]
var = pd.DataFrame(l)
print(var)
```

```python
dct ={
    "course":['python','c++','java','django'],
    "fees" : [1000,200,3000,4582],
    "rank" : [2,3,4,6]
}
r = pd.DataFrame(dct)
print(r)
```

```python
r = pd.DataFrame(dct,columns =["course","rank"])
print(r)
```

```python
dst ={
    "num":[1,2,3,4,5],
    "rank":[2,3,4,5,6],
    "items":[2,6,7,8,2]
}
ar = pd.DataFrame(dst)
print(ar)
```

```python
print(ar["rank"][3])
```

```python
sr1 = pd.Series([1,2,3,4,5])
sr2 = pd.Series([11,12,13,14,15])
sr = {"s1": sr1,"s2": sr2}
print(sr)
```

# Arithematic Operations in pandas

```python
dic ={
"l1" :[1,2,3,4,5,7],
"l2" :[11,12,13,14,15,21]
}
sr = pd.DataFrame(dic)
print(sr)
```

```python
sr["l3"] = sr["l1"] + sr["l2"]
print(sr)
```

```python
sr["cond1"] = sr["l2"] >=15
print(sr)
```

# Insertion and Deletion of Data in Pandas

```python
print(sr)
```

```python
sr.insert(1,"extra",sr["l2"])
print(sr)
```

```python
sr["cut"] = sr["extra"][:3]
print(sr)
```

```python
sr.pop("extra")
```
